```python
#load in relevant libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import recall_score, precision_score
from sklearn.model_selection import cross_validate
from sklearn.model_selection import GridSearchCV
```

```python
#load in our dataset
cancer = pd.read_csv('dataset/wdbc.csv')
cancer
```

```python
cancer['diagnosis'] = cancer['diagnosis'].replace({
    "M" : "Malignant",
    "B":"Benign"
})
cancer
```

```python
#copy our data
standardized_cancer = cancer.copy()

columns_to_exclude = ['id','diagnosis']
columns_to_scale = standardized_cancer.columns.difference(columns_to_exclude)
columns_to_scale
```

```python
scaler = StandardScaler() #initialize our scaler
standardized_cancer[columns_to_scale] = scaler.fit_transform(standardized_cancer[columns_to_scale]) #fit to data
standardized_cancer
```

```python
#set a seed for reproducibility
np.random.seed(1)

#split the data
cancer_train, cancer_test = train_test_split(standardized_cancer, train_size = 0.75, shuffle = True, stratify = standardized_cancer['diagnosis'])
```

```python
cancer_train.info()
```

```python
cancer_test.info()
```

```python
#step 1. initialize our model
knn = KNeighborsClassifier(n_neighbors = 5)
```

```python
#step 2. define our X and y
X = cancer_train[["perimeter_mean","concavity_mean"]]
y = cancer_train['diagnosis']
```

```python
#step 3. fit our model to our data
knn.fit(X,y)
```

```python
#step 4. predict on the test set

cancer_test["predicted"] = knn.predict(cancer_test[["perimeter_mean","concavity_mean"]])
```

```python
cancer_test
```

```python
#compute accuracy
knn.score(
    cancer_test[["perimeter_mean","concavity_mean"]], cancer_test["diagnosis"]
)
```

```python
#confusion matrix

pd.crosstab(
    cancer_test['diagnosis'],
    cancer_test['predicted']
)
```

```python
#precision = "when the model predicted malignant, how often is it actually correct"

precision_score(
    y_true = cancer_test['diagnosis'],
    y_pred = cancer_test['predicted'],
    pos_label = "Malignant"
)
```

```python
#recall = "of all the actual malignant tumors, how many did the model successfully idenitfy"

recall_score(
  y_true = cancer_test['diagnosis'],
  y_pred = cancer_test['predicted'],
  pos_label = "Malignant"  
)
```

```python
#split into subtrain and validation
cancer_subtrain, cancer_validation = train_test_split(cancer_train, train_size = 0.75, stratify = cancer_train["diagnosis"])
```

```python
#fit our model to the subtraining data

knn = KNeighborsClassifier(n_neighbors = 3) #step 1. initialize model
X = cancer_subtrain[["perimeter_mean","concavity_mean"]] #step 2. define X and y
y = cancer_subtrain["diagnosis"]
knn.fit(X,y) #step 3. fit to subtrain data
```

```python
acc = knn.score(
    cancer_validation[["perimeter_mean","concavity_mean"]],
    cancer_validation["diagnosis"]
)
acc
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
X = cancer_train[["perimeter_mean","concavity_mean"]]
y = cancer_train['diagnosis']

returned_dictionary = cross_validate(
    estimator = knn,
    cv = 5,
    X = X,
    y = y
)
```

```python
cv_5_df = pd.DataFrame(returned_dictionary)
cv_5_df #focus on test score column
```

```python
cv_5_metrics = cv_5_df.agg(["mean","sem"])
cv_5_metrics
```

```python
#grid search to find the optimal value of k
parameter_grid = {
    "n_neighbors": range(1, 400, 5)
}
```

```python
#initialized our grid search CV
cancer_tune_grid = GridSearchCV(
    estimator = knn,
    param_grid = parameter_grid,
    cv = 10
)
```

```python
#fit our grid search cv to our training data
cancer_tune_grid.fit(
    cancer_train[["perimeter_mean","concavity_mean"]], 
    cancer_train["diagnosis"]
)
```

```python
#obtain results from grid search
accuracy_grid = pd.DataFrame(cancer_tune_grid.cv_results_)
accuracy_grid
```

```python
# Create the plot
plt.figure(figsize=(10, 6))

# Plot mean test scores with error bars
plt.plot(accuracy_grid['param_n_neighbors'], accuracy_grid['mean_test_score'], '-o', color='blue')

# Add labels and legend
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy estimate')
plt.title('K-Nearest Neighbors Performance')
plt.tight_layout()
plt.show()
```
